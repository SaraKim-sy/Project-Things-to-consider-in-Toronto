```python
import pandas as pd
import os
from pathlib import Path
```

```python
# get current and parent directories using pathlib
curr_dir = Path(os.getcwd())
parent_dir = curr_dir.parents[0]
print('current directory: ' + str(curr_dir) + '\nparent directory:  ' + str(parent_dir))
```

```python
# remove dollar signs and commas
def clean_numeric_str(curr_str):
    if isinstance(curr_str, str):
        return(curr_str.replace('$', '').replace(',', '').replace('%', ''))
    return(curr_str)
```

```python
# crime rate dataset
crime_csv = os.path.join(curr_dir, 'Resources', 'Neighbourhood_Crime_Rates.csv')
crime_df = pd.read_csv(crime_csv)
crime_df = crime_df.drop(['_id','OBJECTID'], axis=1)
crime_df.rename(columns = {'Hood_ID':'Neighbourhood Number'}, inplace = True) 
crime_df.sort_values(by='Neighbourhood Number')
# crime_df['Neighbourhood Number'].dtype
```

```python
# 2016 census data by Toronto neighbourhood available in the City of Toronto Open Data Portal
# https://open.toronto.ca/dataset/neighbourhood-profiles/
census_csv = os.path.join(curr_dir, 'Resources', 'neighbourhoods-profiles-2016-csv-transposed.csv')
census_df = pd.read_csv(census_csv)

# clean up numeric data types
for column in census_df:
    try:
        census_df[column] = census_df[column].apply(clean_numeric_str).astype('float')
    except:
        continue

census_df = census_df.drop(0)
census_df['Neighbourhood Number'] = census_df['Neighbourhood Number'].astype('int64')
census_df.sort_values(by='Neighbourhood Number')
# census_df['Neighbourhood Number'].dtype
```

```python
houseprice_csv = os.path.join(curr_dir, 'Resources', 'wellbeing-toronto-housing.csv')
houseprice_df = pd.read_csv(houseprice_csv)
houseprice_df.rename(columns = {'Neighbourhood Id':'Neighbourhood Number'}, inplace = True) 
houseprice_df['Neighbourhood Number'] = houseprice_df['Neighbourhood Number'].astype('int64')
houseprice_df.sort_values(by='Neighbourhood Number')
# houseprice_df['Neighbourhood Number'].dtype
```

```python
# Combine the data into a single dataset
# Merge on Neighbourhood Number
housing_df = pd.merge(census_df, houseprice_df, on='Neighbourhood Number', how='left')
housing_and_crime_df = pd.merge(housing_df, crime_df, on='Neighbourhood Number', how='left')
housing_and_crime_df
# print(housing_df.dtypes)
# print(housing_and_crime_df.dtypes)
```
