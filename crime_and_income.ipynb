```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
income_df = pd.read_excel('Resources/neighbourhood-income-data-2011.xlsx')
income_df.head()
```

```python
income_df = income_df.loc[(income_df["Category"] == 'Income') & (income_df["Topic"]  == 'Income of households')
& ((income_df["Attribute"] == 'Median household total income $')|(income_df["Attribute"]  == 'Average household total income $'))]
income_df.head()
```

```python
# Did transpose of all cities columns to 1 single column

transposed_df =income_df.loc[:,'Agincourt North':'Yorkdale-Glen Park']
cities_data = transposed_df.transpose().reset_index()
cities_data.columns = ["Cities","Median_Income","Average_Income"]
print(cities_data)
```

```python
# Merge with Crime data

crime_csv = "Crime_rate/Neighbourhood_Crime_Rates.csv"
crime_df = pd.read_csv(crime_csv)
crime_df.head()
```

```python
crime_df = crime_df.loc[:,["Neighbourhood", "Hood_ID", "Assault_Rate_2019", "AutoTheft_Rate_2019", "BreakandEnter_Rate_2019", "Homicide_Rate_2019", "Robbery_Rate_2019", "TheftOver_Rate_2019"]]
```

```python
# Calculate the total crime rate
crime_df["Total Averate Rate 2019"] = round((crime_df["Assault_Rate_2019"] + crime_df["AutoTheft_Rate_2019"] + crime_df["BreakandEnter_Rate_2019"] +
                                       crime_df["Homicide_Rate_2019"] + crime_df["Robbery_Rate_2019"] + crime_df["TheftOver_Rate_2019"]) / 6, 2)
```

```python
# Rename 
crime_df = crime_df.rename(columns={"Assault_Rate_2019" : "Assault Rate",
                                    "AutoTheft_Rate_2019" : "Auto Theft Rate",
                                    "BreakandEnter_Rate_2019" : "Break&Enter Rate",
                                    "Homicide_Rate_2019" :"Homicide Rate",
                                    "Robbery_Rate_2019" : "Robberey Rate",
                                    "TheftOver_Rate_2019" : "Theft Over Rate",
                                    "Total Averate Rate 2019" : "Total Average Rate"}) 
```

```python
# Set index as "Neighbourhood name"
crime_df = crime_df.set_index('Neighbourhood')
```
